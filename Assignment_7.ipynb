```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(102367004)
sales_data = np.random.randint(1000, 5001, size=(12, 4))
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
df = pd.DataFrame(sales_data, columns=categories, index=months)

print("First 5 Rows:")
print(df.head())
print("\nSummary Statistics:")
print(df.describe())

total_sales_category = df.sum()
total_sales_month = df.sum(axis=1)
print("\nTotal Sales per Category:")
print(total_sales_category)
print("\nTotal Sales per Month:")
print(total_sales_month)

avg_growth = df.pct_change().mean()
print("\nAverage Sales Growth per Category:")
print(avg_growth)

df['Total Sales'] = df.sum(axis=1)
df['Growth Rate'] = df['Total Sales'].pct_change() * 100
df['Electronics'] = df['Electronics'] * 0.9

plt.figure(figsize=(10,6))
for col in categories:
    plt.plot(df.index, df[col], label=col)
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(data=df[categories])
plt.title('Sales Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])
print("Original Array:\n", array)

abs_array = np.abs(array)
print("\nElement-wise Absolute Value:\n", abs_array)

flattened = array.flatten()
percentiles_flat = np.percentile(flattened, [25, 50, 75])
print("\nPercentiles of Flattened Array (25th, 50th, 75th):\n", percentiles_flat)

percentiles_col = np.percentile(array, [25, 50, 75], axis=0)
print("\nPercentiles for Each Column (rows: 25th, 50th, 75th):\n", percentiles_col)

percentiles_row = np.percentile(array, [25, 50, 75], axis=1)
print("\nPercentiles for Each Row (rows: 25th, 50th, 75th):\n", percentiles_row)

mean_flat = np.mean(flattened)
median_flat = np.median(flattened)
std_flat = np.std(flattened)
print("\nFlattened Array:\nMean =", mean_flat, ", Median =", median_flat, ", Std Dev =", std_flat)

mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)
print("\nEach Column:\nMean =", mean_col, "\nMedian =", median_col, "\nStd Dev =", std_col)

mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)
print("\nEach Row:\nMean =", mean_row, "\nMedian =", median_row, "\nStd Dev =", std_row)
```

```python
import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
print("Original Array:\n", a)

floor_vals = np.floor(a)
ceil_vals = np.ceil(a)
trunc_vals = np.trunc(a)
rounded_vals = np.round(a)

print("\nFloor Values:\n", floor_vals)
print("\nCeiling Values:\n", ceil_vals)
print("\nTruncated Values:\n", trunc_vals)
print("\nRounded Values:\n", rounded_vals)
```
