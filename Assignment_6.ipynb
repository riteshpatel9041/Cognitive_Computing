```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 400)

y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y1, 'r-', label=r'$y = Mx^2$')
plt.plot(x, y2, 'b--', label=r'$y = M \sin(x)$')

plt.legend()
plt.grid(True)
plt.title(f'Plot of $y = Mx^2$ and $y = M \sin(x)$ for M = {M}')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

subjects = ['Math', 'Science', 'English', 'History', 'Art']
scores = [85, 90, 78, 88, 92]

df = pd.DataFrame({'Subject': subjects, 'Score': scores})

sns.barplot(x='Subject', y='Score', data=df, hue='Subject', palette='husl', dodge=False)

for i, score in enumerate(scores):
    plt.text(i, score + 1, str(score), ha='center')

plt.title('Scores by Subject')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(12345)
data = np.random.randn(50)

fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].plot(np.cumsum(data))
axs[0].set_title('Cumulative Sum')
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Cumulative Value')
axs[0].grid(True)

axs[1].scatter(range(50), data, color='r')
axs[1].set_title('Random Noise')
axs[1].set_xlabel('Index')
axs[1].set_ylabel('Value')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('company_sales_data.csv')

plt.figure(figsize=(10, 5))
sns.lineplot(x=df.index, y=df['total_profit'], marker='o')
plt.title('Total Profit Over Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
for column in df.columns[1:-1]:
    sns.lineplot(x=df.index, y=df[column], label=column)
plt.title('Product Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

df.set_index(df.columns[0]).plot(kind='bar', figsize=(12, 6), colormap='tab10')
plt.title('Sales Data for All Features')
plt.xlabel(df.columns[0])
plt.ylabel('Values')
plt.grid(True, axis='y')
plt.show()
```
