```python

import numpy as np

gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
sum_all = np.sum(gfg)
print("Sum of all elements:", sum_all)

sum_rows = np.sum(gfg, axis=1)
print("Sum of all elements row-wise:", sum_rows)
sum_cols = np.sum(gfg, axis=0)
print("Sum of all elements column-wise:", sum_cols)
```

```python
import numpy as np

array = np.array([10, 52, 62, 16, 16, 54, 453])
sorted_array = np.sort(array)
print("Sorted Array",sorted_array)
sorted_indices = np.argsort(array)
print("sorted_indices",sorted_indices)
smallest_4 = sorted_array[:4]
print("smallest_4",smallest_4)
largest_5 = sorted_array[-5:]
print("largest_5",largest_5)

array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
int_elements = array[array % 1 == 0]
print("Integers",int_elements)
float_elements = array[array % 1 != 0]
print("Float",float_elements)
```

```python
import numpy as np

first_initial = 'A'
last_initial = 'B'
X = ord(first_initial) + ord(last_initial)
sales = np.array([X, X+50, X+100, X+150, X+200])
print("Sales Data:", sales)

tax_rate = ((X % 5) + 5) / 100
sales_after_tax = sales * (1 + tax_rate)
print("Sales After Tax:", sales_after_tax)

discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)
print("Discounted Sales:", discounted_sales)
weeks = 3
weekly_sales = np.vstack([sales] * weeks)
week_factors = np.array([1.00, 1.02, 1.04]).reshape(3, 1)
adjusted_weekly_sales = weekly_sales * week_factors
print("Adjusted Weekly Sales:\n", adjusted_weekly_sales)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y1 = x ** 2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Y = x^2', color="blue")
plt.title('Y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Y = sin(x)', color="green")
plt.title('Y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, y3, label='Y = e^x', color="purple")
plt.title('Y = e^x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, y4, label='Y = log(|x| + 1)', color="brown")
plt.title('Y = log(|x| + 1)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
